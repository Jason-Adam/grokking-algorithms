```python
from collections import deque

def person_is_seller(name):
      return name[-1] == 'm'

graph = {}

#The first level should have five people
graph["me"] = ["david", "eric", "frank", "gary", "henry"]

#and each person in first level should have five connections (they can be
# exclusive to one person or shared with other members on the same level, but all five should be unique, no duplicates)
graph["david"] = ["dylan", "derek", "donna", "don", "darya"]
graph["eric"] = ["evan","ellen","eva","elaine","ed"]
graph["frank"] = ["fanny", "fred","fawn","francis","felix"]
graph["gary"] = ["gabriel", "graham", "garrett", "gregory", "griffin"]
graph["henry"] = ["harold", "hunter", "hugo", "hector", "harrison"]

#Each person in the second level should have five connections (they can be exclusive to one person or shared with other
# members on the same level, but all five should be unique, no duplicates)
graph["dylan"] = ["doug", "drake", "dakota", "dom", "daniel"]
graph["derek"] = ["dax", "darius", "donovan", "damien", "deandre"]
graph["donna"] = ["danielle", "daisy", "diana", "dawn", "delilah"]
graph["don"] = ["denise", "dharma", "daniela", "dominique", "deedee"]
graph["darya"] = ["destiny", "dixie", "dinah", "daryl", "doris"]

graph["evan"] = ["ebony","esther","eba","enna", "elle"]
graph["ellen"] = ["eartha", "eki", "elton", "euretha", "emma"]
graph["eva"] = ["euy", "ekka", "eap","ewe","etta"]
graph["elaine"] = ["eat","ear","emo","epi","equine"]
graph["ed"] = ["eja", "ento", "etty","eya","eee"]

graph["fanny"] = ["fa", "fe", "fo", "fum", "fe"]
graph["fred"] = ["foo", "flea", "flar", "fent", "farr"]
graph["fawn"] = ["fab", "fan", "fat", "for","fee"]
graph["francis"] = ["fua", "fub", "fun", "fud", "fug"]
graph["felix"] = ["fba", "fbb", "fbc", "fbd", "fbe"]

graph["gabriel"] = ["gua, gub", "guc", "gud", "gue"]
graph["graham"] = ["gba, gbb", "gbee", "gbor", "gantt"]
graph["garrett"] = ["glue", "glen", "gar", "gee","gro"]
graph["gregory"] = ["gu", "ga", "gra", "guppy", "geet"]
graph["griffin"] = ["gka", "gkb", "gkc", "gkd", "gke"]

graph["harold"] = ["har", "han", "hat", "harr", "hae"]
graph["hunter"] = ["hee", "her","han", "hott", "hap"]
graph["hugo"] = ["hpa", "hpb", "hpc", "hpd", "hpe"]
graph["hector"] = ["hma", "hmb", "hmc", "hmd", "hme"]
graph["harrison"] = ["hla", "hlb", "hlc", "hld", "hle"]


def search(name):
    search_queue = deque()
    search_queue += graph[name]
    # This array is how you keep track of which people you've searched before.
    searched = []
    while search_queue:
        person = search_queue.popleft()
        # Only search this person if you haven't already searched them.
        if person not in searched:
            if person_is_seller(person):
                print(person + " is a mango seller!")
                return True
            else:
                search_queue += graph[person]
                # Marks this person as searched
                searched.append(person)
    return False

```

In the breadth first search, you search across every level in full before going down to the "next level" - i.e., the neighbors of anyone on the first level. In other words, the neighbors of the first node in breadth-first search are added at the end of the queue, and will be searched after every other item in the first level is searched.(So the order above is Alex, Ben, Candy, Derek... etc.) This differs from depth-first search, where the search goes into as deep into the levels as it can get, before moving back to the next items (i.e. 2nd, 3rd, 4th) on the **First** level. See visual diagram I created above.

I would have expected that, because this is a breadth first search, all of the algos I ran where the word was on the FIRST level would be computationally less taxing than searching for the word on the second level (indices 3-5 in the above dataframe). And, the words searched on the 2nd level would be less taxing than the words on the 3rd level. This isn't what the data illustrated.
 
This relates to big-o notation because as we have discussed, big o notation refers to the performance or complexity of an algorithm. In other words, the execution time of the algorithm -- how long it takes the algo to run. Breadth first search takes  O(number of people + number of edges), since every instance that the search algorithm has to check is O(1). Because Breadth first search and depth first search can be used for different use-cases (i.e. do you need to "dive deep" or "search the entire first level?") you should understand what your goal is. That way, you can choose the computationally cheaper option.
